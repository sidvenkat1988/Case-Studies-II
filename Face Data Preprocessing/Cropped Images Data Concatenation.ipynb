# Concatenating Training Data

```
cropped_images = []
faces_found = []
image = []
label = []

for file in glob.glob('*.npy'):
  cropped_image = np.load(file, allow_pickle=True)
  for i in range(len(cropped_image)):
    cropped_images.append(cropped_image[i])

for file in glob.glob('*.csv'):
  df = pd.read_csv(file)
  for i in df.iloc[:, 1].tolist():
    faces_found.append(i)
  for i in df.iloc[:, 2].tolist():
    image.append(i)
  for i in df.iloc[:, 3].tolist():
    label.append(i) 

print(str(len(cropped_images)) + '- cropped_images')
print(str(len(faces_found)) + '- faces_found')
print(str(len(image)) + '- image')
print(str(len(label)) + '- label')

plt.imshow(cropped_images[-1])
plt.xlabel('image_name - ' + image[-1] + ', label - ' + label[-1])
```
