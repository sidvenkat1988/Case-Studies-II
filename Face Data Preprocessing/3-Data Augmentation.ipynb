```
import pandas as pd
import numpy as np
import os
import time
```

```
import warnings
warnings.filterwarnings('ignore')
```

```
pd.set_option('max_colwidth', -1)
```

```
name = np.load('shortened_name_set.npy')
image_path = np.load('shortened_image_set.npy')
columns = ['name', 'image_path']
df = pd.DataFrame(columns = columns)
df['name'] = name
df['image_path'] = image_path 
df.head()
```

```
df['name'].nunique()
```

```
df['name_count'] = df['name'].map(df['name'].value_counts(dropna=False))
df.head()
```

```
!ls
```

```
top_records = df.sort_values(['name_count'], ascending = False).groupby('name').head(1).head(170)
```

```
top_records
```

```
top_records_name = [i for i in top_records['name']]
print(len(top_records_name))
```

```
df.head()
```

```
top_170_records = []
for i in range(len(df)):
  if df['name'][i] in top_records_name:
    top_170_records.append('yes')
  else:
    top_170_records.append('no')
df['top_170_records'] = top_170_records
df.head()
```

```
final_df = df.loc[df['top_170_records'] == 'yes'].sort_values('name_count', ascending = False).groupby('name').head(9)
```

```
final_df.shape
```

```
final_df.head()
```

```
del final_df['name_count']
del final_df['top_170_records']
df.head()
```

```
final_df.head()
```

```
final_df = final_df.reset_index(drop = True)
final_df.head()
```

**Now this is the final dataframe through which I will be generating all the augmented images.**

# Generating images with ZCA-Whitening, and a rotation range of 40

```
os.mkdir('Augmented_Images')
```

```
%cd Augmented_Images/
```

```
os.mkdir('zca_with_rotation')
%cd zca_with_rotation/
```

```
augmented_path = os.getcwd()
```

```
augmented_path
```

```
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
```

```
datagen = ImageDataGenerator(zca_whitening=True, rotation_range = 40)

for i in range(len(final_df)):
  img = load_img(final_df['image_path'][i])
  x = img_to_array(img)
  x = x.reshape((1,) + x.shape)

  j = 0
  for batch in datagen.flow(x, batch_size = 1, save_to_dir = augmented_path, save_prefix = (final_df['name'][i] + ' - '), save_format = 'jpg'):
    j += 1
    if j > 4:
      break
  
  if len(os.listdir(augmented_path))%4 == 0:
    print(len(os.listdir(augmented_path)))
```
