```
import os
import pandas as pd
import numpy as np
```

```
pd.set_option('max_colwidth', -1)
```

```

labels = []
for i in os.listdir(os.getcwd()):
  #print(os.path.join(os.getcwd(), i))
  labels.append(os.path.join(os.getcwd(), i))
```

```
len(labels)
```

```
os.listdir(labels[0])
```

```
images = []
name = []
for i in labels:
  for j in os.listdir(i):
    #print(os.path.join(i,j))
    images.append(os.path.join(i,j))
    name.append(i)
```

```
images
```

```
columns = ['name', 'images']
df = pd.DataFrame(columns = columns)
df['name'] = name
df['images'] = images
df.head()
```

```
df['name'] = df['name'].str.replace('/content/drive/My Drive/case studies 2/Face Dataset/lfw/', '')
df.head()
```

```
name = []
image_path = []
for i in df['name']:
  name.append(i)
for i in df['images']:
  image_path.append(i)
```
