# Splitting the data

```
%cd ..
```

```
!ls
```

```
augmented_folders = []
for i in os.listdir(os.getcwd()):
  augmented_folders.append(os.path.join(os.getcwd(), i))
augmented_folders
```

```
images = []

for i in augmented_folders:
  for j in os.listdir(i):
    #print(os.path.join(i,j))
    images.append(os.path.join(i,j))
print(len(images))
```

```
columns = ['image_path']
df = pd.DataFrame(columns = columns)
df['image_path'] = images
df.head()
```

```
df['image'] = df['image_path'].str.split('/').str[-1]
df.head()
```

```
df['label'] = df['image'].str.split('-').str[0]
df.head()
```

```
sample_df = df.sample(frac = 1)
sample_df.head()
```

```
sample_df = sample_df.reset_index(drop = True)
sample_df.head()
```

```
sample_df['label2'] = sample_df['image'].str.split('-').str[0]
sanity_check = []
for i in range(len(sample_df)):
  if sample_df['label'][i] == sample_df['label2'][i]:
    sanity_check.append('yes')
  else:
    sanity_check.append('no')
sample_df['sanity_check'] = sanity_check
sample_df.head()
```

```
sample_df.loc[sample_df['sanity_check']== 'yes']
```

```
del sample_df['label2']
del sample_df['sanity_check']
```

```
df = sample_df
df.head()
```

```
df['label_count'] = df['label'].map(df['label'].value_counts(dropna=False))
```

```
df.head()
```

```
df['label_count'].unique()
```

```
final_df = df.sort_values('label', ascending=False).groupby('label').head(311)
```

```
final_df
```

```
training_df = final_df.sort_values('label', ascending=False).groupby('label').head(250)
```

```
image = []
for i in training_df['image']:
  image.append(i)
present = []
for i in final_df['image']:
  if i in image:
    present.append('yes')
  else:
    present.append('no')
final_df['present'] = present
final_df.head()
```

```
final_df['present'].value_counts(dropna=False)
```
