```
import pandas as pd
import numpy as np
import os
import cv2
```

```
pd.set_option('max_colwidth', -1)
```

```
name = np.load('name.npy')
image_path = np.load('image_path.npy')
```

```
columns = ['name', 'image_path']
df = pd.DataFrame(columns = columns)
df['name'] = name
df['image_path'] = image_path
df.head()
```

```
df['count'] = df['name'].map(df['name'].value_counts(dropna=False))
```

```
df.head()
```

```
df['name'].nunique()
```

```
len(df)
```

```
df['count'].nunique()
```

```
df = df.loc[df['count']>5]
```

```
df.head()
```

```
shortened_name_set = []
shortened_image_set = []
for i in df['name']:
  shortened_name_set.append(i)
for i in df['image_path']:
  shortened_image_set.append(i)
```

```
len(shortened_image_set)
```

```
len(shortened_name_set)
```

```
np.save('shortened_name_set', shortened_name_set)
np.save('shortened_image_set', shortened_image_set)
```
