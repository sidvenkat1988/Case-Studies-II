**Extracting speaker data from each data type**

```
speaker_data = []
for i in os.listdir(dev_clean):
  speaker_data.append(os.path.join(dev_clean, i))
for i in os.listdir(dev_other):
  speaker_data.append(os.path.join(dev_other, i))
for i in os.listdir(test_clean):
  speaker_data.append(os.path.join(test_clean, i))
for i in os.listdir(test_other):
  speaker_data.append(os.path.join(test_other, i))
for i in os.listdir(train_clean_100):
  speaker_data.append(os.path.join(train_clean_100, i))
```

```
os.listdir(speaker_data[0])
```

**Extracting chapters belonging to each speaker**

```
chapter_data = []
for i in speaker_data:
  for j in os.listdir(i):
    #print(os.path.join(i,j))
    chapter_data.append(os.path.join(i,j))
```

```
chapter_data[0]
```

```
chapter_data[-1]
```

**Extracting the files of each speaker**

```
files_data = []
for i in chapter_data:
  for j in os.listdir(i):
    #print(os.path.join(i, j))
    files_data.append(os.path.join(i,j))
```
