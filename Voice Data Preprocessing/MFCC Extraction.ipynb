```
import numpy as np
import pandas as pd
import librosa
```

```
import time
```

```
pd.set_option('max_colwidth', -1)
```

```
files = np.load('files_data.npy', allow_pickle=True)
```

```
len(files)
```

```
files[0]
```

```
df = pd.DataFrame(columns = ['files'])
df['files'] = files
df.head()
```

```
df['file_name'] = df['files'].str.split('/').str[-1]
df['speaker'] = df['file_name'].str.split('-').str[0]
df.head()
```

```
len(df)
```

```
df.head()
```

```
mfcc = []
label = []
file_name = []
start = time.time()
for i in range(len(df['files'])):
  try:
    X, sample_rate = librosa.load(df['files'][i], res_type = 'kaiser_fast')
    mfccs = np.mean(librosa.feature.mfcc(y = X, sr = sample_rate, n_mfcc = 40).T, axis = 0)
    mfcc.append(mfccs)
    label.append(df['speaker'][i])
    file_name.append(df['file_name'][i])
    if len(mfcc)%10 == 0:
      print(len(mfcc))
      print(time.time() - start)
  except Exception as e:
    pass
```

```
np.save('mfcc', mfcc, allow_pickle = True)
np.save('label', label, allow_pickle = True)
np.save('file_name', file_name, allow_pickle = True)
```

```
df = pd.DataFrame(columns = ['label'])
df['label'] = label
df['label_counts'] = df['label'].map(df['label'].value_counts(dropna = False))
df.head()
```

```
df = df.loc[df['label_counts']>=100]
```

```
df
```

```
label_list = list(df['label'].unique())
label_list = label_list[:100]
present = []
for i in df['label']:
  if i in label_list:
    present.append('yes')
  else:
    present.append('no')
df['present'] = present
df.loc[df['present'] == 'yes']
```

```
df = df.loc[df['present'] == 'yes']
```

```
df = df.sort_values('label', ascending = False).groupby('label').head(100)
```

```
df
```

```
index = list(df.index)
index
```

```
final_mfcc = []
final_label = []
final_file_name = []

for i in index:
  final_mfcc.append(mfcc[i])
  final_label.append(label[i])
  final_file_name.append(file_name[i])
```
