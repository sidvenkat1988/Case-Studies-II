```
import os
import numpy as np
import pandas as pd
```

```
pd.set_option('max_colwidth', -1)
```

```
files_data = np.load('files_data.npy')
```

```
len(files_data)
```

```
columns = ['files_data']
df = pd.DataFrame(columns = columns)
df['files_data'] = files_data
df.head()
```

```
df.head()
```

```
df = df.loc[df['files_data'].str.contains('flac')]
```

```
len(df)
```

```
df['file_type'] = df['files_data'].str.split('/').str[-4]
```

```
df.head()
```

**Speaker label extraction**

```
df['speaker'] = df['files_data'].str.split('/').str[-3]
df.head()
```

**Unique records of each speaker**

```
df['records'] = df['speaker'].map(df['speaker'].value_counts(dropna=False))
```

```
df.head()
```

```
df['speaker'].nunique()
```

```
df['records'].value_counts(dropna=False)
```

```
df.loc[df['records']>=10]
```

```
df_50 = df.loc[df['records']>= 50]
```

```
df_50
```

```
df_50.sort_values(['speaker'], ascending = False).groupby('speaker').head(50)['file_type'].value_counts(dropna=False)
```

```
df_50.sort_values(['speaker'], ascending = False).groupby('speaker').head(50)['speaker'].value_counts(dropna=False)
```

```
training_data  = df_50.sort_values(['speaker'], ascending = 'False').groupby('speaker').head(50)
```

```
training_data.shape
```

```
validation_data = pd.concat([training_data,df_50]).drop_duplicates(keep=False)
```

```
training_data = training_data.reset_index(drop = True)
```

```
training_data
```

```
validation_data = validation_data.reset_index(drop = True)
```

```
validation_data
```

```
training_file = [i for i in training_data['files_data']]
training_speaker = [i for i in training_data['speaker']]
validation_file = [i for i in validation_data['files_data']]
validation_speaker = [i for i in validation_data['speaker']]
```

```
!pwd
```

```
np.save('training_file', training_file)
np.save('training_speaker', training_speaker)
np.save('validation_file', validation_file)
np.save('validation_speaker', validation_speaker)
```
