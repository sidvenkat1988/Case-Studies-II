```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import librosa
import librosa.display
```

```
data_6000_6500 = np.load('data_6000_6500.npy', allow_pickle=True)
data_6500_7000 = np.load('data_6500_7000.npy', allow_pickle = True)
data_7000_7500 = np.load('data_7000_7500.npy', allow_pickle = True)
data_7500_8000 = np.load('data_7500_8000.npy', allow_pickle = True)
data_8000_8500 = np.load('data_8000_8500.npy', allow_pickle=True)
data_8500_9000 = np.load('data_8500_9000.npy', allow_pickle=True)
data_9000_9500 = np.load('data_9000_9500.npy', allow_pickle=True)
data_9500_10000 = np.load('data_9500_10000.npy', allow_pickle=True)
data_10000_10500 = np.load('data_10000_10500.npy', allow_pickle=True)
data_10500_11000 = np.load('data_10500_11000.npy', allow_pickle=True)
data_11000_rest = np.load('data_11000_rest.npy', allow_pickle=True)
```

```
sr_6000_6500 = np.load('sr_6000_6500.npy')
sr_6500_7000 = np.load('sr_6500_7000.npy')
sr_7000_7500 = np.load('sr_7000_7500.npy')
sr_7500_8000 = np.load('sr_7500_8000.npy')
sr_8000_8500 = np.load('sr_8000_8500.npy')
sr_8500_9000 = np.load('sr_8500_9000.npy')
sr_9000_9500 = np.load('sr_9000_9500.npy')
sr_9500_10000 = np.load('sr_9500_10000.npy')
sr_10000_10500 = np.load('sr_10000_10500.npy')
sr_10500_11000 = np.load('sr_10500_11000.npy')
sr_11000_rest = np.load('sr_11000_rest.npy')
```

```
sr = np.concatenate([sr_6000_6500,
sr_6500_7000,
sr_7000_7500,
sr_7500_8000,
sr_8000_8500,
sr_8500_9000])
```

```
data = np.concatenate([data_6000_6500,
data_6500_7000,
data_7000_7500,
data_7500_8000,
data_8000_8500,
data_8500_9000])
```

```
filename_1 = []
for i in range(len(speaker)):
  start = time.time()
  try:
    fig = plt.figure(figsize=[1,1])
    ax = fig.add_subplot(111)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    S = librosa.feature.melspectrogram(y = data[i][:sr[i]], sr = sr[i])
    librosa.display.specshow(librosa.power_to_db(S, ref = np.max), y_axis = 'linear')
    file = str(speaker[i]) + '- 1 -' + str(i) + '.jpg'
    filename_1.append(file)
    plt.savefig(file, dpi = 500, bbox_inches = 'tight', pad_inches = 0)
    plt.close()
  except Exception as e:
    filename_1.append('error')

  if len(filename_1)%1000==0:
    print(len(filename_1))
    print(len(os.listdir(os.getcwd())))
    print(time.time()-start)
  plt.close() 
```
