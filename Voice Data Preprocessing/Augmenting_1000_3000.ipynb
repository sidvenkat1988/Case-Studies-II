```
import numpy as np
import pandas as pd
import os
import time
import librosa
import librosa.display
import matplotlib.pyplot as plt
```

```
training_speaker = np.load('training_speaker.npy')
```

```
len(training_speaker)
```

```
speaker = training_speaker[1000:3000]
```

```
len(speaker)
```

```
%cd raw_waveforms/
```

```
!ls
```

```
data_1000_1500 = np.load('data_1000_1500.npy', allow_pickle=True)
sr_1000_1500 = np.load('sr_1000_1500.npy')
```

```
data_1500_2000 = np.load('data_1500_2000.npy', allow_pickle=True)
sr_1500_2000 = np.load('sr_1500_2000.npy')
```

```
data_2000_2500 = np.load('data_2000_2500.npy', allow_pickle=True)
sr_2000_2500 = np.load('sr_2000_2500.npy')
```

```
data_2500_3000 = np.load('data_2500_3000.npy', allow_pickle=True)
sr_2500_3000 = np.load('sr_2500_3000.npy')
```

```
data = np.concatenate([data_1000_1500, data_1500_2000, data_2000_2500, data_2500_3000])
sr = np.concatenate([sr_1000_1500, sr_1500_2000, sr_2000_2500, sr_2500_3000])
```

```
filename_1 = []
for i in range(len(speaker)):
  start = time.time()
  try:
    fig = plt.figure(figsize=[1,1])
    ax = fig.add_subplot(111)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    S = librosa.feature.melspectrogram(y = data[i][:sr[i]], sr = sr[i])
    librosa.display.specshow(librosa.power_to_db(S, ref = np.max), y_axis = 'linear')
    file = str(speaker[i]) + '- 1 -' + str(i) + '.jpg'
    filename_1.append(file)
    plt.savefig(file, dpi = 500, bbox_inches = 'tight', pad_inches = 0)
    plt.close()
  except Exception as e:
    filename_1.append('error')

  if len(filename_1)%100==0:
    print(len(filename_1))
    print(time.time()-start)
  plt.close() 
```
