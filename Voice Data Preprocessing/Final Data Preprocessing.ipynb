# Checking the label distribution across folders

```
columns = ['images']
df = pd.DataFrame(columns = columns)
df['images'] = images
df.head()
```

```
df['label_count'] = df['label'].map(df['label'].value_counts(dropna=False))
df.head()
```

```
testing_df['image_name'] = testing_df['images'].str.split('/').str[-1]
testing_df.head()
```

```
testing_df['label'] = testing_df['image_name'].str.split('-').str[0]
testing_df.head()
```
