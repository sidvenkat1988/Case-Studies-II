```
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
import time
import librosa
import librosa.display
```

```
data_3000_3500 = np.load('data_3000_3500.npy', allow_pickle=True)
data_3500_4000 = np.load('data_3500_4000.npy', allow_pickle = True)
data_4000_4500 = np.load('data_4000_4500.npy', allow_pickle = True)
data_4500_5000 = np.load('data_4500_5000.npy', allow_pickle=True)
data_5000_5500 = np.load('data_5000_5500.npy', allow_pickle=True)
data_5500_6000 = np.load('data_5500_6000.npy', allow_pickle=True)
```

```
sr_3000_3500 = np.load('sr_3000_3500.npy')
sr_3500_4000 = np.load('sr_3500_4000.npy')
sr_4000_4500 = np.load('sr_4000_4500.npy')
sr_4500_5000 = np.load('sr_4500_5000.npy')
sr_5000_5500 = np.load('sr_5000_5500.npy')
sr_5500_6000 = np.load('sr_5500_6000.npy')
```

```
%cd ..
```

```
!ls
```

```
training_speaker = np.load('training_speaker.npy')
```

```
speaker = training_speaker[3000:6000]
print(len(speaker))
```

```
data = np.concatenate([data_3000_3500, data_3500_4000, data_4000_4500, data_4500_5000, data_5000_5500, data_5500_6000])
print(len(data))
```

```
sr = np.concatenate([sr_3000_3500, sr_3500_4000, sr_4000_4500, sr_4500_5000, sr_5000_5500, sr_5500_6000])
print(len(sr))
```

```
filename_1 = []
for i in range(len(speaker)):
  start = time.time()
  try:
    fig = plt.figure(figsize=[1,1])
    ax = fig.add_subplot(111)
    ax.axes.get_xaxis().set_visible(False)
    ax.axes.get_yaxis().set_visible(False)
    ax.set_frame_on(False)
    S = librosa.feature.melspectrogram(y = data[i][:sr[i]], sr = sr[i])
    librosa.display.specshow(librosa.power_to_db(S, ref = np.max), y_axis = 'linear')
    file = str(speaker[i]) + '- 1 -' + str(i) + '.jpg'
    filename_1.append(file)
    plt.savefig(file, dpi = 500, bbox_inches = 'tight', pad_inches = 0)
    plt.close()
  except Exception as e:
    filename_1.append('error')

  if len(filename_1)%300==0:
    print(len(filename_1))
    print(len(os.listdir(os.getcwd())))
    print(time.time()-start)
  plt.close() 
```
